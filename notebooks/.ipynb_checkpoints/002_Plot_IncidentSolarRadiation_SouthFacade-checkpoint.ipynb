```python
import pandas as pd
from iertools.read import read_sql
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = pd.read_csv('../data/Is.csv',index_col=0,parse_dates=True)
```

```python
df['time'] = df.index.hour + df.index.minute/60
df['day']    = df.index.day_of_year
df.head()
```

```python
sns.relplot(data=df,
            x='time',
            y='Is',
            alpha=0.05,
            hue='day',
            kind='line', 
            ci=100)
```
